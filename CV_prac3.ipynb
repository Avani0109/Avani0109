# Prac A

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('/content/OIP.webp', cv2.IMREAD_GRAYSCALE)
```

```python
avg_filter = cv2.blur(img, (5,5))
```

```python
weighted_filter = cv2.GaussianBlur(img, (5,5), 0)
```

```python
median_filter = cv2.medianBlur(img, 5)
```

```python
titles = ['Original Image', 'Average Filter', 'weighted (Gaussian filter)', 'median filter']
images = [img, avg_filter, weighted_filter, median_filter]
```

```python
plt.figure(figsize=(12, 6))

plt.subplot(1,4,1)
plt.imshow(images[0])
plt.title(titles[0])
plt.axis('off')

plt.subplot(1,4,2)
plt.imshow(images[1])
plt.title(titles[1])
plt.axis('off')

plt.subplot(1,4,3)
plt.imshow(images[2])
plt.title(titles[2])
plt.axis('off')

plt.subplot(1,4,4)
plt.imshow(images[3])
plt.title(titles[3])
plt.axis('off')

plt.tight_layout()
plt.show()




```

```python
plt.figure(figsize=(12, 6))
for i in range(4):


  plt.subplot(1,4,i+1)
  plt.imshow(images[i], cmap='gray')
  plt.title(titles[i])
  plt.axis('off')

plt.tight_layout()


```

# Prac 1A part-2

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('/content/OIP (1).webp', cv2.IMREAD_GRAYSCALE)
```

```python
sobel_x = cv2.Sobel(img, cv2.CV_64F, 1, 0, ksize = 3)
sobel_y = cv2.Sobel(img, cv2.CV_64F, 0, 1, ksize = 3)
sobel_combined = cv2.magnitude(sobel_x, sobel_y)
```

```python
sobel_combined = cv2.convertScaleAbs(sobel_combined)
```

```python
laplacian = cv2.Laplacian(img, cv2.CV_64F, ksize = 3)
laplacian = cv2.convertScaleAbs(laplacian)
```

```python
titles = ['Original Color', 'Sobel operator', 'Laplacian Operator']
images = [img, sobel_combined, laplacian]
```

```python
plt.figure(figsize=(12,6))
for i in range(3):
  plt.subplot(1,3,i+1)
  plt.imshow(images[i])
  plt.title(titles[i])
  plt.axis('off')

plt.tight_layout()
```

# Prac 3C

```python
# Median filter to remove salt and paper noise
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('/content/sp_img_gray_noise_heavy.png', cv2.IMREAD_GRAYSCALE)
```

```python
median = cv2.medianBlur(img, 5)
```

```python
plt.subplot(1,2,1)
plt.imshow(img, cmap = 'gray')
plt.title('Original Image')
plt.axis('off')

plt.subplot(1,2,2)
plt.imshow(median, cmap='gray')
plt.title('Median Filtered Image')
plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
# Bilateral filter
bilateral = cv2.bilateralFilter(img, d=9, sigmaColor=75, sigmaSpace=75)

plt.subplot(1,2,1)
plt.imshow(img, cmap = 'gray')
plt.title('Original Image')
plt.axis('off')

plt.subplot(1,2,2)
plt.imshow(bilateral, cmap='gray')
plt.title('Bilateral Filtered Image')
plt.axis('off')

plt.tight_layout()
plt.show()
```
