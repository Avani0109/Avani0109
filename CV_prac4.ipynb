# Prac 4A

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('/content/OIP (1).webp', cv2.IMREAD_GRAYSCALE)
```

```python
sobelx = cv2.Sobel(img, cv2.CV_64F, 1, 0, ksize=3)
sobely = cv2.Sobel(img, cv2.CV_64F, 0, 1, ksize=3)
```

```python
mag = cv2.magnitude(sobelx, sobely)
```

```python
mag = cv2.convertScaleAbs(mag)
```

```python
plt.figure(figsize=(10, 5))

plt.subplot(1,3,1)
plt.imshow(sobelx, cmap='gray')
plt.title('Sobel X')

plt.subplot(1,3,2)
plt.imshow(sobely, cmap='gray')
plt.title('Sobel Y')

plt.subplot(1,3,3)
plt.imshow(mag, cmap='gray')
plt.title('Magnitude')

plt.tight_layout()
plt.show()
```

# Prac 4B

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow
from google.colab import files
```

```python
uploaded = files.upload()
```

```python
filename = list(uploaded.keys())[0]
img = cv2.imread(filename, cv2.IMREAD_GRAYSCALE)
```

```python
prewittx = np.array([[-1, 0, 1], [-1, 0, 1], [-1, 0, 1]])
prewitty = np.array([[-1, -1, -1], [0,0,0], [1,1,1]])
```

```python
gx = cv2.filter2D(img, -1, prewittx)
gy = cv2.filter2D(img, -1, prewitty)
```

```python
mag = cv2.convertScaleAbs(np.sqrt(gx.astype(np.float32)**2+gy.astype(np.float32)**2))
```

```python
plt.figure(figsize = (12,6))
titles = ['Prewitt X', 'Prewitt Y', 'Magnitude']
images = [gx, gy, mag]

for i in range(3):
  plt.subplot(1,3,i+1)
  plt.imshow(images[i], cmap='gray')
  plt.title(titles[i])
  plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files
```

```python
files = files.upload()
```

```python
filename = list(uploaded.keys())[0]
img = cv2.imread(filename)
```

```python
img = cv2.imread(filename, cv2.IMREAD_GRAYSCALE)
```

```python
blurred = cv2.GaussianBlur(img, (5,5), sigmaX=1.0)
```

```python
laplacian = cv2.Laplacian(blurred, cv2.CV_64F, ksize = 3)
lapabs = cv2.convertScaleAbs(laplacian)
```

```python
plt.figure(figsize=(10,5))
plt.subplot(1,3,1)
plt.imshow(img, cmap='gray')
plt.title('Original Image')
plt.axis('off')

plt.subplot(1,3,2)
plt.imshow(blurred, cmap='gray')
plt.title('Blurred Image')
plt.axis('off')

plt.subplot(1,3,3)
plt.imshow(lapabs, cmap='gray')
plt.title('Laplacian')
plt.axis('off')

plt.tight_layout()
plt.show()
```
