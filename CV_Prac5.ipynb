```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files
```

```python
uploaded = files.upload()
```

```python
# Step 3: Read the image
imgpath = list(uploaded.keys())[0]
img = cv2.imread(imgpath)
```

```python
# Step 4: Convert the image to grayscale
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
# step 5: Apply thresholding
ret, thresh1 = cv2.threshold(gray, 0, 255, cv2.THRESH_BINARY+cv2.THRESH_OTSU)
```

```python
fig, (ax1, ax2) = plt.subplots(1,2, figsize=(12, 5))
ax1.imshow(gray, cmap='gray')
ax1.set_title('Grayscale Image')
ax1.axis('off')

ax2.imshow(thresh1, cmap='gray')
ax2.set_title('Threshold Image')
ax2.axis('off')

plt.show()
```

# Prac B

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files
```

```python
uploaded = files.upload()
```

```python
imgpath = list(uploaded.keys())[0]
img = cv2.imread(imgpath)
```

```python
b, g, r = cv2.split(img)
```

```python
ret_b, thresh_b = cv2.threshold(b, 0, 255, cv2.THRESH_BINARY+cv2.THRESH_OTSU)
ret_g, thresh_g = cv2.threshold(g, 0, 255, cv2.THRESH_BINARY+cv2.THRESH_OTSU)
ret_r, thresh_r = cv2.threshold(r, 0, 255, cv2.THRESH_BINARY+cv2.THRESH_OTSU)
```

```python
thresh = cv2.merge((thresh_b, thresh_g, thresh_r))
```

```python
fig, (ax1, ax2) = plt.subplots(1,2, figsize=(12, 5))
ax1.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
ax1.set_title('Original Image')
ax1.axis('off')

ax2.imshow(cv2.cvtColor(thresh, cv2.COLOR_BGR2RGB))
ax2.set_title('Thresholded Image')
ax2.axis('off')

plt.show()

print('Threshold for Red channel: ', ret_r)
print('Threshold for Green channel: ', ret_g)
print('Threshold for Blue channel: ', ret_b)
```

# Prac 1D

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files
```

```python
uploaded = files.upload()
```

```python
imgpath = list(uploaded.keys())[0]
img = cv2.imread(imgpath)
```

```python
imgfloat = img.astype(np.float32)
```

```python
c = 255/(np.log(1+np.max(imgfloat)))
log = c*np.log(1+imgfloat)
```

```python
log_trans = c*np.log(1+imgfloat)
```

```python
log_trans = np.array(log_trans, dtype=np.uint8)
```

```python
fig, (ax1, ax2) = plt.subplots(1,2,figsize = (12,6))
ax1.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
ax1.set_title('Original Image')
ax1.axis('off')

ax2.imshow(cv2.cvtColor(log_trans, cv2.COLOR_BGR2RGB))
ax2.set_title('Log Transformed Image')
ax2.axis('off')

plt.show()
```
